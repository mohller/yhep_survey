```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
```

### Choose the palette

```python
# This is the colorset used in the BMBAF presentation
original_colorset = ['#72B4DE', '#F19F85', '#7AE17A', '#E4E577', '#B6D1B3', '#DFCDB5', '#FEF1EB', '#66CBC6', '#C8D6E5', '#E8C1E5', '#BDBDF7']

cmap = cm.get_cmap('coolwarm', 5)
yesno_colorset = [cmap(k) for k in range(1, 4)] + ['grey', 'tan']

cmap = cm.get_cmap('Greens', 7)
one_to_five_colorset = [cmap(k) for k in range(1, 6)] + ['lightgrey',]

colorset = original_colorset
```

# Diggest of Survey by yHEP

```python
table = pd.read_csv('results-survey548259_all_replies.csv')

dropped_cols = []
#dropping all NaN columns
for col in table.columns:
    try:
        if np.all(np.isnan(table[col])):
            dropped_cols.append(col)
    except:
        continue
        
table.drop(dropped_cols, inplace=True, axis=1)
print('Dropped columns: ', dropped_cols)

table
```

# Reproducing the plots in the presentation

The scripts below reproduce the figures in the RECFA visit [slides](https://docs.google.com/presentation/d/1DnN_1kzu5QPZZQkMM-CPSQU3EJpafiP17_E7hzEcy6c/edit#slide=id.g11a4a1ead1d_0_17)

## Distribution by Section 

```python
polled_by_section = table['Which yHEP section do you see yourself in?']

labels = polled_by_section.unique()
values = []

for val in labels:
    values.append(sum(polled_by_section == val))

values, labels = np.array(values), np.array(labels)
sorting_sequence = np.argsort(values)[::-1]

values = values[sorting_sequence]
labels = labels[sorting_sequence]

def my_autopct(pct):
    return f'{pct:.0f}%' if pct > 5 else ''

plt.pie(values, labels=labels, counterclock=False, autopct=my_autopct,
        startangle=90, textprops={'fontsize':15, 'wrap':True},
        colors=colorset);
```

## Duration of doctoral studies

```python
PhDduration1 = table['How long are your doctoral / PhD studies planned to last?']
PhDduration2 = table['How long did it take you to complete your doctorate / PhD?']

labels = ['< 3 years', '3 years', '4 years', '5 years', '> 5 years', 'Other', "I don't know               ", ]
values = []

for duration in labels:
#     values.append(sum(PhDduration1 == duration) + sum(PhDduration2 == duration))
    values.append(sum(PhDduration2 == duration))

plt.pie(values, labels=labels, counterclock=False, startangle=90, textprops={'fontsize':15, 'wrap':True},
        colors=colorset);

plt.gca().set_title('Duration of doctoral/PhD studies', fontsize=20, fontweight='bold')
```

## General happiness in different categories

```python
work_situation = table['We are interested to hear about your work situation [Are you happy with your work situation?]']
worklife_balance = table['We are interested to learn more about your work conditions [Are you happy with your current work-life balance?]']
familywork_balance = table['How would you rate your family-work balance?']

figure_labels = ['Work situation', 'Work-life balance', 'Family-Work balance']

f, axes = plt.subplots(1, 3, figsize=(14, 4))

for ax, series, fl in zip(axes, [work_situation, worklife_balance, familywork_balance], figure_labels):

    subser = series.value_counts()
    subser.sort_index(inplace=True)
    if fl != 'Work situation':
        
        subser = subser.append(pd.Series({'nan' : 1}))

    ax.set_title(fl, fontsize=20, fontweight='bold')
    ax.pie(subser, labels=subser.keys(), counterclock=False, startangle=90, textprops={'fontsize':15, 'wrap':True},
        colors=one_to_five_colorset);
    
cbax = plt.legend(loc=(1.08, 0.24), fontsize=16)
plt.annotate(xy=(0.8, -1.25), text='*Leftmost pie wrongly counts \nnan values as one, yielding 22 \nanswers instead of 21.', fontsize=16)
```

## What is appreciated and what not

```python
ScientificFreedom = table['We are interested to hear about your work situation [Do you have sufficient scientific freedom?]']

values = ScientificFreedom.value_counts().sort_index()

def my_autopct(pct):
    return f'{pct:.0f}%' if pct > 5 else ''

plt.pie(values, labels=values.keys(), counterclock=False, autopct=my_autopct,
        startangle=90, textprops={'fontsize':15, 'wrap':True}, colors=one_to_five_colorset);

plt.gca().set_title('Amount of scientific freedom', fontsize=20, fontweight='bold')
```

## Contract situation

```python
f, axes = plt.subplots(1, 2, figsize=(18, 4))

plt.sca(axes[0])

funding_source = table['How is your current position funded?']

values = funding_source.value_counts()
values = values.append(pd.Series({'No answer' : 1, #sum(funding_source.isna()), # wrongly assuming a smaller number of no answer
                                  'Max Planck institute' : 0}))

original_keys = values.keys()
reordered_keys = [original_keys[l] for l in [0, 2, 1, 7, 4, 10, 8, 3, 6, 5, 9]]

values = values.reindex(reordered_keys)

def my_autopct(pct):
    return f'{pct:.0f}%' if pct > 5 else ''

plt.pie(values, labels=values.keys(), counterclock=False, autopct=my_autopct,
        startangle=90, textprops={'fontsize':15, 'wrap':True}, colors=colorset, labeldistance=None);

plt.gca().set_title('Source of funding', fontsize=20, fontweight='bold')
plt.legend(loc=(1., 0.), fontsize=12)


plt.sca(axes[1])

unemployment_periods = table['A few questions on your contract situation: [Have you ever encountered short periods of unemployment because of non-matching contracts?]']

values = unemployment_periods.value_counts()
values = values.append(pd.Series({'No answer' : 1 #sum(unemployment_periods.isna())
                                 }))

original_keys = values.keys()
reordered_keys = [original_keys[l] for l in [1, 2, 0, 3]]

values = values.reindex(reordered_keys)

def my_autopct(pct):
    return f'{pct:.0f}%' if pct > 5 else ''

plt.pie(values, labels=values.keys(), counterclock=False, autopct=my_autopct,
        startangle=90, textprops={'fontsize':15, 'wrap':True}, colors=yesno_colorset, labeldistance=None);

plt.gca().set_title('Encountered periods \nof unemployment', fontsize=20, fontweight='bold')
plt.legend(loc=(1., 0.5), fontsize=12)
```

## Mental Health

```python
f = plt.figure(figsize=(20, 4), dpi=120)

ax1 = plt.subplot2grid((1,8), (0,0), colspan=5)
ax2 = plt.subplot2grid((1,8), (0,5), colspan=2)
axes = [ax1, ax2]

plt.sca(axes[0])

overwork = table['Do you think you overwork?']

values = overwork.value_counts()
values = values.append(pd.Series({'No answer' : sum(overwork.isna()), #sum(overwork.isna()), # wrongly assuming a smaller number of no answer
                                 # 'I do not want to answer' : 0
                                 }))

original_keys = values.keys()
reordered_keys = [original_keys[l] for l in [1, 0, 2, 3, 4]]

values = values.reindex(reordered_keys)

def my_autopct(pct):
    return f'{pct:.0f}%' if pct > 5 else ''

plt.pie(values, labels=values.keys(), counterclock=False, autopct=my_autopct,
        startangle=90, textprops={'fontsize':15, 'wrap':True}, colors=yesno_colorset, labeldistance=None);

plt.gca().set_title('Impression to overwork', fontsize=20, fontweight='bold')
plt.legend(loc=(1., 0.4), fontsize=12)


plt.sca(axes[1])

symptoms = table['Do you have the impression that you have or are developing physical symptoms or an impact on your health from overwork? ']

values = symptoms.value_counts()
values = values.append(pd.Series({'No answer' : 1 #sum(unemployment_periods.isna())
                                 }))

original_keys = values.keys()
reordered_keys = [original_keys[l] for l in [1, 0, 3, 2, 4, 5]]

values = values.reindex(reordered_keys)


plt.bar(range(len(values)), values, color=colorset);
plt.xticks(range(6), ['1', '2', '3', '4', '5', 'No answer'])
plt.ylabel('Number of replies')

plt.gca().set_title('Impression of an impact on health (incl.\nphysical symptoms) from overwork', fontsize=20, fontweight='bold')
```

## Job opportunities in the field

```python
f = plt.figure(figsize=(20, 4), dpi=120)

ax1 = plt.subplot2grid((1,8), (0,0), colspan=5)
ax2 = plt.subplot2grid((1,8), (0,5), colspan=2)
axes = [ax1, ax2]

plt.sca(axes[0])

jobs_academia = table['A few general questions to rate [How good do you see your chances for a permanent job in academia?]']

values = jobs_academia.value_counts()
values = values.append(pd.Series({'No answer' : sum(overwork.isna()), #sum(overwork.isna()), # wrongly assuming a smaller number of no answer
                                  }))

original_keys = values.keys()
reordered_keys = [original_keys[l] for l in [4, 3, 2, 1, 0, 5]]

values = values.reindex(reordered_keys)


plt.pie(values, labels=values.keys(), counterclock=False,
        startangle=90, textprops={'fontsize':15, 'wrap':True}, colors=one_to_five_colorset, labeldistance=None);

plt.gca().set_title('Percieved opportunities\nfor a permanent job', fontsize=20, fontweight='bold')
plt.legend(loc=(1., 0.4), fontsize=12)

plt.gca().text(-.01, -.9, 'In academia', rotation=35, fontsize=15)

# ---- Inset plot
ax3 = f.add_axes([.164, .3, .4, .4])
plt.sca(ax3)

jobs_outside_academia = table['A few general questions to rate [How good do you see your chances for a permanent job outside academia?]']
values = jobs_outside_academia.value_counts()

plt.pie(values, labels=values.keys(), counterclock=False,
        startangle=90, textprops={'fontsize':15, 'wrap':True}, colors=one_to_five_colorset, labeldistance=None);

plt.gca().text(-.8, -.6, 'Outside\nacademia', rotation=35, fontsize=15)

plt.sca(axes[1])

vals_dict = {}

for kt in table.keys():
    if kt.startswith('If you leave'):
        vals_dict[kt[kt.index('[')+1:kt.index(']')]] = sum(table[kt] == 'Yes')
vals_dict['Other'] = 4 # differs from slides, don't know why


from textwrap import wrap
for kt, c in zip(vals_dict.keys(), colorset):
    lab = '\n'.join(wrap(kt, 20))
    plt.plot([], [], ls='', marker='s', ms=11, color=c, label=lab)


plt.bar(range(len(vals_dict.keys())), vals_dict.values(), color=colorset);
plt.xticks(range(11), ('',)*11)
plt.ylabel('Number of replies')

plt.gca().set_title('Main reasons to leave', fontsize=20, fontweight='bold')
plt.legend(loc=(1.01, -.05))
```

## Gender balance

```python
disadvantage = table['Have you had the impression that you are at a career-related disadvantage due to gender, sexual orientation, ethnicity, disability, etc.?']
gender = table['What is your gender?']

values = disadvantage[gender == 'Female'].value_counts()
values = values.append(pd.Series({'No answer' : 1}))


original_keys = values.keys()
reordered_keys = [original_keys[l] for l in [1, 2, 0, 3, 4]]
values = values.reindex(reordered_keys)


f, ax = plt.subplots(dpi=120)

plt.pie(values, labels=values.keys(), counterclock=False, labeldistance=None, 
        startangle=90, textprops={'fontsize':15, 'wrap':True}, colors=yesno_colorset);

plt.gca().set_title('Impression of a career-related disadvantage due to\ngender, sexual orientation, ethnicity, disability, etc.', fontsize=12, fontweight='bold')
plt.legend(loc=(1., 0.4), fontsize=12)

plt.gca().text(.1, -.85, 'Women', rotation=35, fontsize=15)


f.add_axes([.313, .3, .4, .4])

values = disadvantage[gender == 'Male'].value_counts()
values = values.append(pd.Series({'No answer' : 1}))


original_keys = values.keys()
reordered_keys = [original_keys[l] for l in [1, 2, 0, 3, 4]]
values = values.reindex(reordered_keys)


plt.pie(values, labels=values.keys(), counterclock=False, labeldistance=None, shadow=True,
        startangle=90, textprops={'fontsize':15, 'wrap':True}, colors=yesno_colorset, wedgeprops={'linewidth':5});

plt.gca().text(-.3, -.5, 'Men', rotation=35, fontsize=15)
```
